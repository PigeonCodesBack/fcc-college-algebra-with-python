# Code

Set up a proportion:<br>
$\frac{n1}{d1}=\frac{n2}{d2}$

```python
# Put a zero in for the unknown value
n1 = 1
d1 = 2
n2 = 4
d2 = 0

# Do not edit under this line

if n2==0:
    print(f"{n1} / {n2} = x / {d2}")
    answer = d2 * n1 / d1
    print(f"x = {answer}")

if d2==0:
    print(f"{n1} / {n2} = {n2} / x")
    answer = n2 * d1 / n1
    print(f"x = {answer}")
```
